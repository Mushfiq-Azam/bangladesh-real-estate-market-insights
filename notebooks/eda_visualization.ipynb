```python
# Ensure plotly is installed in this Jupyter environment
%pip install plotly

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_cleaned_sell_only = pd.read_csv('cleaned_dhaka_real_estate.csv')
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
```

```python
# 1. Distribution of Property Prices
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned_sell_only['Price_BDT'], bins=20, kde=True)
plt.title('Distribution of Property Prices (in BDT)')
plt.xlabel('Price (BDT)')
plt.ylabel('Frequency')
plt.show()
```

```python
# 2. Price vs Area (scatter plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_cleaned_sell_only, x='Area_sqft', y='Price_BDT', hue='Location', palette='tab20', size='Bedroom', sizes=(50, 200))
plt.title('Price vs Area of Properties')
plt.xlabel('Area (sqft)')
plt.ylabel('Price (BDT)')
plt.legend(title='Location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 3. Heatmap of Price Correlation with other variables
correlation = df_cleaned_sell_only[['Area_sqft', 'Price_BDT', 'Bedroom', 'Bathroom', 'Floor']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# 4. Boxplot of Price distribution by Location
plt.figure(figsize=(14, 6))
sns.boxplot(data=df_cleaned_sell_only, x='Location', y='Price_BDT')
plt.title('Price Distribution by Location')
plt.xlabel('Location')
plt.ylabel('Price (BDT)')
plt.xticks(rotation=90)
plt.show()
```

```python
# 5. Distribution of Properties by Bedroom Count
plt.figure(figsize=(10, 6))
sns.countplot(data=df_cleaned_sell_only, x='Bedroom', palette='viridis')
plt.title('Distribution of Properties by Bedroom Count')
plt.xlabel('Bedroom Count')
plt.ylabel('Number of Properties')
plt.show()
```
