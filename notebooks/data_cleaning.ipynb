```python
import pandas as pd
```

```python
file_path = 'dhaka_real_estate.csv'
df = pd.read_csv(file_path)
```

```python
df.columns = df.columns.str.strip()
```

```python
df['Area_sqft'] = pd.to_numeric(df['Area_sqft'], errors='coerce')
```

```python
df['Price_BDT'] = df['Price_BDT'].replace({'BDT': '', 'Crore': 'e7', 'Lakh': 'e5'}, regex=True)
df['Price_BDT'] = pd.to_numeric(df['Price_BDT'], errors='coerce')
```

```python
df = df.dropna(subset=['Price_BDT', 'Area_sqft'])
```

```python
df = df.drop_duplicates()
```

```python
# Filter the dataset to keep only the rows where the 'For' column is 'Sell'
df_cleaned_sell_only = df[df['For'] == 'Sell']

# Save the cleaned dataset with only "Sell" properties
df_cleaned_sell_only.to_csv('cleaned_dhaka_real_estate.csv', index=False)

# Display the first few rows of the cleaned data to verify
df_cleaned_sell_only.head()
```
